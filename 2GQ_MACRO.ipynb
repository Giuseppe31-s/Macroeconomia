##1. Explique o conceito de NAIRU.###
####Relação de fixação dos salários.
> A NAIRU em alguns casos também é usada no sentido de taxa natural de    desemprego.É a taxa de desemprego necessária para manter a taxa de inflação constante. (Non Accelerating Inflation Rate of Unemployment).Para Phelps e Friedman a NAIRU é a taxa  de  desemprego  natural,  para  a  qual  a  economia converge.   A   NAIRU   é   associada   a   um produtopotencial  máximo,  o  nível  mais  alto  de  produção  que pode  ser  sustentado  no  longo  prazo.

>Vamos entender o que realmente significa a taxa natural de desemprego.
$$W = PF(u,z)$$
onde:
* W é salário nominal agregado
* P é nível esperado dos preços
* u é desemprego
* z é todas as variáveis que podem afetar a fixação dos salários


> Esse é o modelo básico que explica a estrutura do salário agregado, vamos encontrar a fórmula do desemprego.

>Isolando a função desemprego:
$$ \frac{W}{P}=F(u,z)$$

> O salário real tem uma relação negativa com a taxa de desemprego, claro quanto maior a taxa de desemprego, menor vai ser o poder de barganha, logo se inclue um menor salário. Essa relação entre salário real e a taxa de desemprego é chamada de relação de fixação dos salários.

####Relação de fixação dos preços.
 > Para determinar o preço é usada a seguinte equação:

 $$P = (1+m)W$$


> Divindo ambos os lados pelo salário nominal:

$$\frac{P}{W} = 1+m$$

>Vamos inverter os dois lados da equação para obter o salário real.

$$\frac{W}{P} = \frac{1}{1+m}$$

> A equação nos informa que: as decisões de fixação determinam o salário real pago pelas empresas. Essa equação é chamada de relaçao de fixação de preços e não depende do desemprego. Vamos montar um simples modelo para entendermos melhor.






















> A diversas formas de utilização desse conceito, a mais básica é traçar uma média móvel da taxa de desemprego para ter uma noção de onde  a taxa natural de desemprego se estabelece. No exemplo, a seguir, foram usado dados atualizados da Pesquisa Nacional por Amostra de Domicílios Contínua(PNAD)




```python
#Importando bibliotecas
import pandas as pd
import math
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = list(range(5,55,5))
WP = [8]*10
WS = [20,17,15,12,10,9,8,7,5,4.8]
L =[]
for i in WS:
  F = i - 1
  L.append(F)
```

```python

plt.figure(figsize=(9,7))
plt.plot(x,WP)
plt.plot(x,L)
plt.axvline(30,c='r',lw=1,ls=':',ymin=0,ymax=0.3)
plt.xticks([30],['$u_{n}$'],fontsize=15)
plt.yticks([8],['$\\frac{1}{1+m}$'],fontsize=15)
plt.ylabel('Salário real W/P',fontdict={'fontsize':10,'fontweight':'bold'})
plt.xlabel('Taxa de desemprego',fontdict={'fontsize':10,'fontweight':'bold'})
plt.text(29.9,8,'o A')
plt.text(50.4,8,'PS')
plt.text(50.4,3.6,'WS')
plt.text(35,8.3,'Relação de fixação de preço')
plt.text(35,3.2,'Relação de fixação de salários');
```

```python
#importando dados
url = 'https://raw.githubusercontent.com/Giuseppe31-s/Macroeconomia/main/dados/dados_desemprego.csv'
taxa_desemprego = pd.read_csv(url,sep = ';',decimal = ',')
```

```python
#tratando dados
taxa_desemprego['Trimestre Móvel'] = pd.date_range(start = '02/01/2012', periods=len(taxa_desemprego), freq='M')
taxa_desemprego.rename(columns={'%':'taxa_desemprego','Trimestre Móvel':'data'},inplace=True)
```

```python
#criando média móvel, com base 5
taxa_desemprego['média_móvel'] = taxa_desemprego['taxa_desemprego'].rolling(5).mean()
```

```python
#preparando gráfico
sns.set()
plt.figure(figsize=(10,6))
plt.plot(taxa_desemprego['data'],taxa_desemprego['média_móvel'], label = 'Taxa natural de desemprego' )
plt.plot(taxa_desemprego['data'],taxa_desemprego['taxa_desemprego'], label = 'Taxa de desemprego')
plt.legend()
plt.xlabel('Data', fontdict={'fontsize':12})
plt.ylabel("Taxa em %", fontdict = {'fontsize':12})
plt.title('Taxa Natural de Desemprego', fontdict={'fontsize':18,'fontweight':'bold'})
plt.figtext(0.80,0.15,'Fonte: PNAD',fontdict={'fontsize':8,'fontweight':'bold'});
```

## 2. Por que, segundo Solow, as economias que não apresentam progresso tecnológico, não crescem no longo prazo?

>Para Solow a tecnologia é um fator de produção, é literalmente uma variável do cálculo, logo quanto maior ela for, maior vai ser  a produção e vice-versa. Para entendermos o conceito, vamos atrás da equação. 

> Função de produção

>$Y = F(K,L)$

> Adicionando a variável tecnologia

> $Y = F(K,AL)$

> Onde Y representa o produto agregado, L representa o fator de trabalho, K o fator de produção e A representa a tecnologia vigente.

>esse conceito é também conhecido como **resíduo de Solow** é a parte do crescimento econômico decorrente do progresso tecnológico
