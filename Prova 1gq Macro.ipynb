Separando os pacotes

```python
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = list(range(5,51,5))
LM = [5]*10
LM2 = [10]*10
IS = [30,26,24,20,13,10,7,6,5,4]
mkon = [30,44]
```

```python
plt.rcParams['figure.figsize'] = (12,8)
plt.title('Economia em risco',fontsize = 16)
plt.plot(x,LM,lw =2, label = 'Economia normal',ls =':')
plt.plot(x,LM2,lw =2, label = 'Economia de risco',ls =':')
plt.plot(x,IS, lw = 2,label = 'Produção')
plt.legend()
plt.xlabel('Produção', fontsize =15)
plt.ylabel('Juros', fontsize =15)
plt.xticks([30,44],['LM',"LM'"])
plt.grid(True, axis='x')
plt.text(30,10,'o A')
plt.text(44,5,"o A'")

```

transformando object em dados úteis 

```python
inflação.Data = pd.to_datetime(inflação.Data, format='%d-%m-%y')
```

Criando uma nova coluna com os juros reias

```python
inflação['Juros_reais'] = selic.selic - inflação.inflacao
```

Plotando gráfico

```python
plt.rcParams['figure.figsize'] = (12,8)
plt.title('Juros de Pessoas Físicas, inflação, Selic e Juros Reais',fontsize = 16)
plt.plot(inflação.Data,inflação.inflacao,lw =2, label = ' Inflação')
plt.plot(inflação.Data,inflação.Juros_reais,lw =2, label = 'Juros Reais', c = 'r')
plt.plot(inflação.Data,selic.selic, lw = 2,label = 'Selic',c = 'g')
plt.plot(inflação.Data,taxa.taxa,lw =2, label = 'Juros Pessoa Física',ls ='solid')
plt.ylabel("Juros",fontsize = 15)
plt.xlabel('Anos',fontsize = 15)
plt.legend(loc = 'lower left')
```
